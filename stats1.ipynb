```python
def cal_variance(data):
  #1. mean of data
  n =len(data)
  mean = sum(data)/n
  #2. check deviation
  dev = [(x-mean)**2 for x in data]
  #3. cal variance
  var = sum(dev)/(n)

  return var

cal_variance(data_copy)
```

```python
#write a code for the Mean Ab dev
def mean_ab(data):
  n=len(data)
  mean=sum(data)/n
  dev=[abs(x-mean) for x in data]

  mad = sum(dev)/(n)
  return mad

mean_ab(data_copy)

#seperate the code base on pop and sample
```
